```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn import metrics
import pandas as pd

# Step 1: Load dataset
df = pd.read_csv('/content/week6/Country-data (2).csv')

# Step 3: Compute correlation only on numeric columns
corr_matrix = df.select_dtypes(include=['number']).corr()
print(corr_matrix)
```

```python
df.info()
X= df['health']
Y=df['income']
import numpy as np
data=np.array(list(zip(X,Y)))
data
X
```

```python
# same thing using index location
xx = df.iloc[:,[3,5]]
xx
```

using elbow method to find out the optimal number of clusters


```python
from yellowbrick.cluster import KElbowVisualizer
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,12)).fit(xx)
visualizer.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=0).fit(xx)
kmeans.labels_
```

```python
#We assign the labels to each row in dataframe.
df["Clus_km"] = kmeans.labels_
df.head(15)
```

```python
kmeans.cluster_centers_
```

```python
#Get each cluster size,
from collections import Counter
Counter(kmeans.labels_)
```

```python
from sklearn import metrics
clusters = kmeans.labels_.tolist()

ClusterCentroid=kmeans.cluster_centers_
clustervalidation=metrics.silhouette_score(xx,clusters, metric='euclidean')
print(clustervalidation)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(data=xx, x="health", y="income", hue=kmeans.labels_)
plt.show()
```

```python
sns.scatterplot(data=xx, x="health", y="income", hue=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],
            marker="X", c="r", s=80, label="centroids")
plt.legend()
plt.show()
```

Assuming for new value of income and health


```python
newdata=[[85.89,30400],[10,10000],[15,1700],[3,6000]]
d=pd.DataFrame(newdata,columns=['health','income'])
#d=[[15,1700]]
predictedClusters=kmeans.predict(d)
print(len(predictedClusters))
print(d)
print(predictedClusters)
```

```python
import pandas as pd
from sklearn.cluster import KMeans

# Load dataset
df = pd.read_csv('/content/sample_data/diabetesN (8).csv')

# Select only the columns for clustering
X = df[['BMI', 'Glucose']]

# Fit KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)  # fit only on BMI and Glucose

# New data (same features as used in fit)
newdata = [[85.89, 30400], [10, 10000], [15, 1700], [3, 6000]]
d = pd.DataFrame(newdata, columns=['BMI','Glucose'])

# Predict clusters
predictedClusters = kmeans.predict(d)
print(predictedClusters)
```
