```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Step 1: Transactions
transactions = [
    ['Mango', 'Onion', 'Nintendo', 'Key-chain', 'Eggs', 'Yo-yo'],
    ['Doll', 'Onion', 'Nintendo', 'Key-chain', 'Eggs', 'Yo-yo'],
    ['Mango', 'Apple', 'Key-chain', 'Eggs'],
    ['Mango', 'Umbrella', 'Corn', 'Key-chain', 'Yo-yo'],
    ['Corn', 'Onion', 'Key-chain', 'Ice-cream', 'Eggs']
]

# Step 2: Convert to one-hot encoded dataframe
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)

# Step 3: Find frequent itemsets (min_support = 0.6)
frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)
print("Frequent Itemsets:\n", frequent_itemsets)

# Step 4: Generate strong association rules (min_confidence = 0.8)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.8)
print("\nStrong Association Rules:\n", rules[['antecedents','consequents','support','confidence','lift']])
```
