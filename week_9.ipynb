```python
from itertools import combinations
# Dataset
transactions = [
    {"Mango", "Onion", "Nintendo", "Key-chain", "Eggs", "Yo-yo"},
    {"Doll", "Onion", "Nintendo", "Key-chain", "Eggs", "Yo-yo"},
    {"Mango", "Apple", "Key-chain", "Eggs"},
    {"Mango", "Umbrella", "Corn", "Key-chain", "Yo-yo"},
    {"Corn", "Onion", "Key-chain", "Ice-cream", "Eggs"}
]

min_sup = 0.6      # Minimum support
min_conf = 0.8     # Minimum confidence
num_trans = len(transactions)
min_sup_count = min_sup * num_trans

# Function to calculate support count
def support_count(itemset):
    return sum(1 for t in transactions if itemset.issubset(t))

# Generate candidate itemsets of size k
def generate_candidates(prev_freq_sets, k):
    candidates = []
    prev_list = list(prev_freq_sets)
    for i in range(len(prev_list)):
        for j in range(i+1, len(prev_list)):
            union_set = prev_list[i] | prev_list[j]
            if len(union_set) == k and union_set not in candidates:
                candidates.append(union_set)
    return candidates

# Step 1: Frequent 1-itemsets
items = {item for t in transactions for item in t}
L1 = [{item} for item in items if support_count({item}) >= min_sup_count]
frequent_itemsets = [L1]

# Step 2: Frequent k-itemsets
k = 2
current_L = L1
while current_L:
    candidates = generate_candidates(current_L, k)
    Lk = [c for c in candidates if support_count(c) >= min_sup_count]
    if Lk:
        frequent_itemsets.append(Lk)
        current_L = Lk
        k += 1
    else:
        break

# Display frequent itemsets
print("FREQUENT ITEMSETS:")
for level in frequent_itemsets:
    for itemset in level:
        print(f"{set(itemset)} | Support = {support_count(itemset)/num_trans:.2f}")

print("\nSTRONG ASSOCIATION RULES (Confidence ≥ 0.8):")

# Function to calculate confidence
def confidence(A, B):
    return support_count(A | B) / support_count(A)

# Function to calculate lift
def lift(A, B):
    return (support_count(A | B)/num_trans) / ((support_count(A)/num_trans)*(support_count(B)/num_trans))

# Generate strong association rules
for level in frequent_itemsets[1:]:
    for itemset in level:
        length = len(itemset)
        for r in range(1, length):
            for left in combinations(itemset, r):
                A = set(left)
                B = itemset - A
                conf = confidence(A, B)
                if conf >= min_conf:
                    sup = support_count(itemset)/num_trans
                    lf = lift(A, B)
                    print(f"{A} → {B} | Support = {sup:.2f}, Confidence = {conf:.2f}, Lift = {lf:.2f}")
```
