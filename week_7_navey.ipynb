```python
import pandas as pd

# Dataset
data = {
    'Weather': ['Rain','Snow','Clear','Clear','Snow','Rain','Rain','Snow','Clear','Clear'],
    'Road': ['Bad','Average','Bad','Good','Good','Average','Good','Bad','Good','Bad'],
    'Traffic': ['High','Normal','Light','Light','Normal','Light','Normal','High','High','High'],
    'Engine': ['No','Yes','No','Yes','No','No','No','No','Yes','Yes'],
    'Accident': ['Yes','Yes','No','No','No','No','No','Yes','No','Yes']
}

df = pd.DataFrame(data)

# New scenario to predict
X = ('Rain', 'Good', 'Normal', 'No')

def naive_bayes_predict(df, X):
    classes = df['Accident'].unique()
    results = {}

    for cls in classes:
        subset = df[df['Accident'] == cls]
        prior = len(subset) / len(df)

        likelihood = 1
        for feature, value in zip(['Weather','Road','Traffic','Engine'], X):
            count = len(subset[subset[feature] == value])
            n_values = df[feature].nunique()  # number of possible values for feature
            likelihood *= (count + 1) / (len(subset) + n_values)  # Laplace smoothing

        results[cls] = prior * likelihood

    # Normalize probabilities
    total = sum(results.values())
    for cls in results:
        results[cls] /= total

    return results

# Predict probabilities
probabilities = naive_bayes_predict(df, X)
print("Posterior Probabilities:", probabilities)

# Predicted class
predicted_class = max(probabilities, key=probabilities.get)
print("Predicted Class:", predicted_class)
```

```python
import pandas as pd

# Play Tennis Dataset
data = {
    'outlook':   ['sunny','sunny','overcast','rain','rain','rain','overcast','sunny',
                  'sunny','rain','sunny','overcast','overcast','rain'],
    'temp':      ['hot','hot','hot','mild','cool','cool','cool','mild',
                  'cool','mild','mild','mild','hot','mild'],
    'humidity':  ['high','high','high','high','normal','normal','normal','high',
                  'normal','normal','normal','high','normal','high'],
    'windy':     [False, True, False, False, False, True, True, False,
                  False, False, True, True, False, True],
    'play':      ['no','no','yes','yes','yes','no','yes','no',
                  'yes','yes','yes','yes','yes','no']
}

df = pd.DataFrame(data)

# Function to compute Naive Bayes Posterior Probability
def naive_bayes(df, X):
    classes = df['play'].unique()
    results = {}

    for cls in classes:
        subset = df[df['play'] == cls]
        prior = len(subset) / len(df)

        likelihood = 1
        for feature, value in zip(['outlook','temp','humidity','windy'], X):
            count = len(subset[subset[feature] == value])
            total = len(subset)
            likelihood *= count / total  # no smoothing needed for this dataset

        results[cls] = prior * likelihood

    # Normalize
    total_prob = sum(results.values())
    for cls in results:
        results[cls] /= total_prob

    return results


# ---- Prediction for X = (sunny, hot, high, false) ----
X = ('sunny', 'hot', 'high', False)
prob_X = naive_bayes(df, X)
print("Posterior Probabilities for X:", prob_X)
print("Predicted Class for X:", max(prob_X, key=prob_X.get))

# ---- Prediction for X' = (overcast, cool, high, true) ----
X2 = ('overcast', 'cool', 'high', True)
prob_X2 = naive_bayes(df, X2)
print("\nPosterior Probabilities for X':", prob_X2)
print("Predicted Class for X':", max(prob_X2, key=prob_X2.get))
```

```python
import pandas as pd

# Loan Dataset
data = {
    'EmploymentStatus': ['Employed', 'Unemployed', 'Employed', 'Employed', 'Unemployed'],
    'CreditHistory': ['Good', 'Bad', 'Good', 'Bad', 'Good'],
    'IncomeLevel': ['High', 'Low', 'Medium', 'Medium', 'Low'],
    'LoanApproved': ['Yes', 'No', 'Yes', 'No', 'Yes']
}

df = pd.DataFrame(data)

# Naive Bayes function for categorical data
def naive_bayes(df, X):
    classes = df['LoanApproved'].unique()
    results = {}

    for cls in classes:
        subset = df[df['LoanApproved'] == cls]
        prior = len(subset) / len(df)

        likelihood = 1
        for feature, value in zip(['EmploymentStatus','CreditHistory','IncomeLevel'], X):
            count = len(subset[subset[feature] == value])
            total = len(subset)
            # Laplace smoothing (avoid zero probabilities)
            unique_vals = df[feature].nunique()
            likelihood *= (count + 1) / (total + unique_vals)

        results[cls] = prior * likelihood

    # Normalize to get probabilities
    total_prob = sum(results.values())
    for cls in results:
        results[cls] /= total_prob

    return results


# Question 1
X1 = ('Employed', 'Good', 'Medium')
prob_X1 = naive_bayes(df, X1)
print("Posterior Probabilities for X1:", prob_X1)
print("Prediction for X1:", max(prob_X1, key=prob_X1.get))

# Question 2
X2 = ('Unemployed', 'Bad', 'Low')
prob_X2 = naive_bayes(df, X2)
print("\nPosterior Probabilities for X2:", prob_X2)
print("Prediction for X2:", max(prob_X2, key=prob_X2.get))
```

```python
import pandas as pd

# Dataset
data = [
    ['Yes', 'No', 'Yes', 'No', 1],
    ['No', 'Yes', 'No', 'No', 0],
    ['Yes', 'Yes', 'Yes', 'Yes', 1],
    ['No', 'No', 'No', 'No', 0],
    ['Yes', 'No', 'No', 'Yes', 1],
    ['No', 'Yes', 'Yes', 'No', 0],
]

df = pd.DataFrame(data, columns=['Fever', 'Cough', 'Fatigue', 'TravelHistory', 'DiseaseDiagnosis'])

# Convert categorical features to numeric
for col in ['Fever', 'Cough', 'Fatigue', 'TravelHistory']:
    df[col] = df[col].map({'Yes': 1, 'No': 0})

# Features and target
X = df[['Fever', 'Cough', 'Fatigue', 'TravelHistory']]
y = df['DiseaseDiagnosis']

# Naive Bayes predictor function
def naive_bayes_predict(X_train, y_train, X_test):
    probs = {}
    classes = y_train.unique()
    for cls in classes:
        cls_data = X_train[y_train == cls]
        prior = len(cls_data) / len(X_train)
        likelihood = 1
        for col in X_train.columns:
            value = X_test[col]
            # Laplace smoothing
            feature_prob = (cls_data[col].sum() + 1) / (len(cls_data) + 2)
            if value == 0:
                feature_prob = 1 - feature_prob
            likelihood *= feature_prob
        probs[cls] = prior * likelihood
    # Normalize
    total = sum(probs.values())
    for cls in probs:
        probs[cls] /= total
    return probs

# Test patients
test_patient_1 = {'Fever': 1, 'Cough': 0, 'Fatigue': 1, 'TravelHistory': 0}
test_patient_2 = {'Fever': 0, 'Cough': 1, 'Fatigue': 0, 'TravelHistory': 0}

# Get probabilities
prob_patient1 = naive_bayes_predict(X, y, test_patient_1)
prob_patient2 = naive_bayes_predict(X, y, test_patient_2)


for i, probs in enumerate([prob_patient1, prob_patient2], 1):
    print(f"Patient {i} probabilities:")
    print(f"Positive (1) = {probs[1]:.3f} → {probs[1]*100:.1f}%")
    print(f"Negative (0) = {probs[0]:.3f} → {probs[0]*100:.1f}%")
    prediction = 1 if probs[1] > probs[0] else 0
    print(f"Prediction: {'Positive' if prediction == 1 else 'Negative'}\n")
```
