```python
# Step 1: Import libraries
import nltk
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Download stopwords
nltk.download('stopwords')

# Step 2: Sample labeled dataset (larger so the model can learn)
data = [
    ("I love this product", "positive"),
    ("This is terrible", "negative"),
    ("I am not sure about this", "neutral"),
    ("The service was amazing and prompt", "positive"),
    ("I hate waiting in long lines", "negative"),
    ("The food was okay, nothing special", "neutral"),
    ("What a wonderful experience!", "positive"),
    ("The product broke after one day", "negative"),
    ("I feel indifferent about the new update", "neutral"),
    ("Customer support was very helpful", "positive"),
    ("The website is confusing and slow", "negative"),
    ("Itâ€™s fine, not great but not bad", "neutral")
]

# Step 3: Split into features (X) and labels (y)
X, y = zip(*data)

# Step 4: Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

# Step 5: Vectorize text using CountVectorizer with English stopwords
from nltk.corpus import stopwords
vectorizer = CountVectorizer(stop_words=stopwords.words('english'))
X_train_counts = vectorizer.fit_transform(X_train)
X_test_counts = vectorizer.transform(X_test)

# Step 6: Train Naive Bayes classifier
classifier = MultinomialNB()
classifier.fit(X_train_counts, y_train)

# Step 7: Make predictions
predictions = classifier.predict(X_test_counts)

# Step 8: Evaluate model
accuracy = accuracy_score(y_test, predictions)
conf_matrix = confusion_matrix(y_test, predictions)
class_report = classification_report(y_test, predictions, zero_division=1)

# Step 9: Print results
print(f"Accuracy: {accuracy:.2f}")
print("\nConfusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(class_report)
```
